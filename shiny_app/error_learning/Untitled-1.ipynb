```python
import pandas as pd
```

```python
data = pd.read_csv("behavioral_data\\error_data.csv")
```

```python
data = data.iloc[:, :8]
data = data.dropna(subset=['participant'])
data.loc[data['phase'] == 'learn', 'phase'] = 'study'
data[:35]
```

```python
# Splitting the DataFrame into two based on 'phase'
study_df = data[data['phase'] == 'study']
test_df = data[data['phase'] == 'test']

# Reshaping the study DataFrame
study_df = study_df.rename(columns={'response': 'study_response', 'rt': 'study_rt'})
study_df = study_df.drop(['phase', 'correct'], axis=1)

# Reshaping the test DataFrame
test_df = test_df.rename(columns={'response': 'test_response', 'rt': 'test_rt'})
test_df = test_df.drop(['phase', 'correct'], axis=1)

# Merging the DataFrames
final_df = pd.merge(study_df, test_df, on=['participant', 'cue', 'target', 'condition'], how='outer')

final_df
```

```python
final_df.to_csv('behavioral_data\\formatted_data.csv')
```
